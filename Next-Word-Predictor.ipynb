```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Assuming content has been tokenized and input_sequences has been created

# Step 1: Get max_len from padded sequences
max_len = max([len(x) for x in input_sequences])

# Step 2: Pad the input sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

# Prepare X and y
x = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]

# Step 3: Build the model
model = Sequential()

# Set input_dim to len(tokenizer.word_index) + 1 (to include the padding token)
# Set input_length to max_len (max sequence length after padding)
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=max_len))

# Add LSTM layer
model.add(LSTM(150))

# Output Dense layer with the size of vocabulary (len(tokenizer.word_index) + 1)
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Explicitly build the model
model.build(input_shape=(None, max_len))

# Print the summary
model.summary()
```

```python
import time
text = "what is the fee"

for i in range(5):
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded_token_text = pad_sequences([token_text],maxlen=56,padding='pre')
    pos = np.argmax(model.predict(padded_token_text))
    for word,index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
            time.sleep(1)
```

```python
# One-hot encode the target labels
y_one_hot = to_categorical(y, num_classes=4994)  # Adjust num_classes based on your dataset

# Evaluate the model on training data
loss, accuracy = model.evaluate(x, y_one_hot, verbose=1)

# Print the results
print(f"Model Loss: {loss}")
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.callbacks import TensorBoard
import time
from sklearn.base import BaseEstimator, ClassifierMixin
from tensorflow.keras.utils import to_categorical  # Import this to one-hot encode y
import os

# Ensure proper UTF-8 encoding in stdout
os.environ['PYTHONIOENCODING'] = 'utf-8'

# Read text data
file_path = 'Demo.txt'  # Your text file
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Remove non-ASCII characters (optional, for data cleaning)
content = ''.join([char for char in content if ord(char) < 128])

# Tokenize text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts([content])
word_index = tokenizer.word_index

# Prepare input sequences for training
input_sequences = []
for sentence in content.split('\n'):
    sequence = tokenizer.texts_to_sequences([sentence])[0]
    if sequence:
        for i in range(1, len(sequence)):
            input_sequences.append(sequence[:i+1])

max_len = max([len(x) for x in input_sequences])
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

# Prepare X and Y for training
x = padded_input_sequences[:, :-1]
y = padded_input_sequences[:, -1]

# One-hot encode the target labels (y)
y = to_categorical(y, num_classes=len(word_index) + 1)  # +1 because word_index starts at 1

# Define the model creation function
def create_model(embedding_dim=100, lstm_units=150, dropout_rate=0.2, learning_rate=0.001):
    model = Sequential()
    model.add(Embedding(input_dim=len(word_index) + 1, output_dim=embedding_dim, input_length=max_len-1))
    model.add(LSTM(lstm_units, dropout=dropout_rate, recurrent_dropout=dropout_rate))
    model.add(Dense(len(word_index) + 1, activation='softmax'))  # Output layer has size equal to vocab size
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Custom Keras model wrapper for use with RandomizedSearchCV
class KerasModelWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, embedding_dim=100, lstm_units=150, dropout_rate=0.2, learning_rate=0.001):
        self.embedding_dim = embedding_dim
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.learning_rate = learning_rate
        self.model = None

    def fit(self, X, y):
        self.model = create_model(self.embedding_dim, self.lstm_units, self.dropout_rate, self.learning_rate)
        self.model.fit(X, y, epochs=10, validation_split=0.2, verbose=1)
        return self

    def predict(self, X):
        return self.model.predict(X)

    def score(self, X, y):
        loss, accuracy = self.model.evaluate(X, y, verbose=0)
        return accuracy

# Hyperparameter grid for RandomizedSearchCV
param_dist = {
    'embedding_dim': [100, 200],
    'lstm_units': [150, 200, 250],
    'dropout_rate': [0.2, 0.3, 0.4],
    'learning_rate': [0.001, 0.0005, 0.0001]
}

# Create and fit the RandomizedSearchCV
keras_model = KerasModelWrapper()
random_search = RandomizedSearchCV(estimator=keras_model, param_distributions=param_dist,
                                   n_iter=5, cv=3, verbose=1, n_jobs=-1, error_score='raise')

# Perform RandomizedSearchCV
random_search_result = random_search.fit(x, y)

# Get best parameters and model
best_params = random_search_result.best_params_
best_model = random_search_result.best_estimator_

print("Best Parameters:", best_params)

# Save the best model
best_model.model.save('best_model.h5')

# Model Evaluation
loss, accuracy = best_model.score(x, y)
print(f"Model Loss: {loss}")
print(f"Model Accuracy: {accuracy}")

# Predict the next word (for testing the prediction process)
text = "what is the fee"
for i in range(5):
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded_token_text = pad_sequences([token_text], maxlen=max_len-1, padding='pre')
    pos = np.argmax(best_model.model.predict(padded_token_text))
    for word, index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
            time.sleep(1)
```
